```python
import plotly.graph_objects as go

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/farhanajaved/Chainlink_Oracle/main/datset_18dec_updated.csv")



#fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default

fig = px.violin(df, y="1_min")
fig1 = px.violin(df, y="2_min")
fig2 = px.violin(df, y="3_min")

fig.update_layout(
    xaxis_title='1 minute',
    yaxis_title='Amount LINK',
    yaxis_exponentformat="none",
     template='simple_white',
     font=dict(
                #family="",
                size=12,
                color="black",
               
                
                 )
      ,width=330
    ,height=330)

fig1.update_layout(
    xaxis_title='2 minutes',
    yaxis_title='Amount LINK',
    template='simple_white',
    
     font=dict(
                #family="",
                size=12,
                color="black",
               
                
                 )
      ,width=330
    ,height=330)

fig2.update_layout(
    xaxis_title='3 minutes',
    yaxis_title='Amount LINK',
     template='simple_white',
     
     font=dict(
                #family="",
                size=12,
                color="black",
               
                
                 )
      ,width=330
    ,height=330)

fig.update_traces(box_visible=True, meanline_visible=True)

fig1.update_traces(box_visible=True, meanline_visible=True)

fig2.update_traces(box_visible=True, meanline_visible=True, fillcolor='lightseagreen', line_color='black')


fig.show()
fig1.show()
fig2.show()
```
